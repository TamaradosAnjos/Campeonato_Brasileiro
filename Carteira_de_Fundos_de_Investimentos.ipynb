```python
import pandas as pd
import requests
import zipfile

pd.options.display.float_format = '{:.4f}'.format
```

```python
#ano = "2021"
#mes = "12"
url = 'https://dados.cvm.gov.br/dados/FII/DOC/INF_MENSAL/DADOS/inf_mensal_fii_2024'
#url = f'https://dados.cvm.gov.br/dados/FI/DOC/INF_DIARIO/DADOS/inf_diario_fi_{ano}{mes}.zip'
```

```python
download = requests.get(url)

with open(f"inf_diario_fi_{ano}{mes}.zip", "wb") as arquivo_cvm:

    arquivo_cvm.write(download.content)

arquivo_zip = zipfile.ZipFile(f"inf_diario_fi_{ano}{mes}.zip")
```

```python
dados_fundos = pd.read_csv(arquivo_zip.open(arquivo_zip.namelist()[0]), sep = ";", encoding = 'ISO-8859-1')

dados_fundos
```

```python
dados_cadastro = pd.read_csv('https://dados.cvm.gov.br/dados/FI/CAD/DADOS/cad_fi.csv',
                             sep = ";", encoding = 'ISO-8859-1')

dados_cadastro = dados_cadastro[['CNPJ_FUNDO', 'DENOM_SOCIAL']]

dados_cadastro = dados_cadastro.drop_duplicates()

dados_cadastro
```

```python
data_inicio_mes = (dados_fundos['DT_COMPTC'].sort_values(ascending = True).unique())[0]
data_fim_mes = (dados_fundos['DT_COMPTC'].sort_values(ascending = True).unique())[-1]


dados_fundos_filtrado = dados_fundos[(dados_fundos['DT_COMPTC'].isin([data_inicio_mes, data_fim_mes]))]


dados_fundos_filtrado
```

```python
base_final = pd.merge(dados_fundos_filtrado, dados_cadastro, how = "left",
                      left_on = ["CNPJ_FUNDO"], right_on = ["CNPJ_FUNDO"])

base_final = base_final[['CNPJ_FUNDO', 'DENOM_SOCIAL', 'DT_COMPTC', 'VL_QUOTA', 'VL_PATRIM_LIQ', 'NR_COTST']]

base_final
```

```python
fundo_az_quest = base_final[base_final['DENOM_SOCIAL'].str.contains("AZ QUEST MASTER", na = False)]

fundo_az_quest
```

```python
print(fundo_az_quest)
```

```python
# Filtrando os dados do fundo específico
fundo_az_quest_fic = fundo_az_quest[fundo_az_quest['CNPJ_FUNDO'] == "23.686.472/0001-45"]

# Verifique se o DataFrame filtrado não está vazio
if fundo_az_quest_fic.empty:
    print("Nenhum dado encontrado para o CNPJ especificado.")
else:
    # Imprimindo os valores de VL_QUOTA para verificar
    primeiro_valor = fundo_az_quest_fic['VL_QUOTA'].iloc[0]
    ultimo_valor = fundo_az_quest_fic['VL_QUOTA'].iloc[-1]
    print(f"Primeiro valor da cota: {primeiro_valor}")
    print(f"Último valor da cota: {ultimo_valor}")

    # Cálculo do retorno
    retorno_dez = ("Retorno fundo: " +
                   str(((ultimo_valor / primeiro_valor - 1) * 100).round(2))
                   + "%")

    print(retorno_dez)
```

```python
fundo_az_quest_fic = fundo_az_quest[(fundo_az_quest['CNPJ_FUNDO'] == "23.686.472/0001-45")]

retorno_dez = ("Retorno fundo: " +
            str(((fundo_az_quest_fic['VL_QUOTA'].iloc[-1]/fundo_az_quest_fic['VL_QUOTA'].iloc[0] - 1)
                 * 100).round(2))
               + "%")

retorno_dez
```
